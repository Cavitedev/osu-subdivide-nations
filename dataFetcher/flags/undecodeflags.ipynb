# Undecode flags

Decode percent-encoded Wikimedia flag URLs in the public flags JSON file.

```python
import json
import urllib.parse
from pathlib import Path

repo_root = Path.cwd()
flags_path = repo_root / 'extension' / 'public' / 'flags.json'
out_flags_path = repo_root / 'dataFetcher' / 'flags' / 'out' / 'flags.json'
out_flags_path.parent.mkdir(parents=True, exist_ok=True)

if not flags_path.exists():
    raise FileNotFoundError(f'Flags file not found: {flags_path}')

with open(flags_path, 'r', encoding='utf-8') as f:
    flags = json.load(f)

updated_count = 0
for country in flags.values():
    if 'regions' not in country:
        continue
    for region in country['regions'].values():
        if 'flag' not in region:
            continue
        flag = region['flag']
        if isinstance(flag, str):
            decoded = urllib.parse.unquote(flag)
            if decoded != flag:
                region['flag'] = decoded
                updated_count += 1
                print(f'Decoded: {flag} -> {decoded}')

with open(flags_path, 'w', encoding='utf-8') as f:
    json.dump(flags, f, ensure_ascii=False, indent=4)

with open(out_flags_path, 'w', encoding='utf-8') as f:
    json.dump(flags, f, ensure_ascii=False, indent=4)

print(f'Updated {updated_count} flag URLs')
print('Saved to', flags_path)
print('Saved to', out_flags_path)
```
